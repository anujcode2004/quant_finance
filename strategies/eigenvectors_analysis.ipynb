## üç≥ Part 1: The Simple Story - Why Do We Care?

### Real-World Example:

Imagine you have 50 friends, and you want to know: **"Do they hang out together?"**

- Some friends always go to parties together (high correlation)
- Some friends never see each other (low correlation)
- Some friends actively avoid each other (negative correlation)

**In finance, it's the same idea!**
- We have 50 stocks
- We want to know: **"Do they move together?"**
- If Apple stock goes up, does Microsoft also go up? (correlation)
- If Oil stocks go up, do Tech stocks go down? (negative correlation)

**Why does this matter?**
- If all your stocks move together, you're not really diversified
- If some stocks move opposite, you have better diversification
- Understanding this helps you build better portfolios!

### The Problem We're Solving:

**Question:** Out of 50 stocks, which ones move together? Can we simplify this?

**Answer:** Yes! Using something called "PCA" (Principal Component Analysis), we can:
- Find the main "patterns" in how stocks move
- Reduce 50 stocks to maybe 5-10 important patterns
- Understand what drives the market

**Think of it like this:**
- Instead of tracking 50 individual stocks
- We find the 5-10 "themes" that explain most of what's happening
- Like: "Tech stocks move together" is one theme
- "Energy stocks move together" is another theme

## üìö Part 2: What is Correlation? (Simple Explanation)

### In Plain English:

**Correlation** = How much two things move together

**Examples:**
- **Correlation = +1.0**: Perfect together! When one goes up, the other ALWAYS goes up
- **Correlation = +0.5**: Pretty similar. When one goes up, the other USUALLY goes up
- **Correlation = 0.0**: No relationship. They move independently
- **Correlation = -0.5**: Opposite! When one goes up, the other USUALLY goes down
- **Correlation = -1.0**: Perfect opposite! When one goes up, the other ALWAYS goes down

### Real Stock Example:

**Apple (AAPL) vs Microsoft (MSFT):**
- Both are tech companies
- They probably move together (high positive correlation, maybe 0.7-0.8)
- When tech is doing well, both do well
- When tech crashes, both crash

**Apple (AAPL) vs Oil Company (XOM):**
- Completely different industries
- They probably don't move together much (low correlation, maybe 0.1-0.2)
- Tech and oil are independent

### The Math (Simple Version):

If we have two stocks, A and B:
- We look at how their prices change over time
- If they change in the same direction ‚Üí positive correlation
- If they change in opposite directions ‚Üí negative correlation
- If they change randomly ‚Üí zero correlation

**That's it!** Correlation is just measuring "do they move together?"

## üìä Part 3: What is Variance? (Simple Explanation)

### In Plain English:

**Variance** = How much something "spreads out" or "varies"

### Simple Example:

**Example 1: Low Variance**
- Stock price: $100, $101, $100, $101, $100
- Very stable! Not much variation
- **Low variance** = Predictable, doesn't change much

**Example 2: High Variance**
- Stock price: $100, $120, $80, $130, $90
- Very volatile! Changes a lot
- **High variance** = Unpredictable, changes a lot

### In Our Context:

When we say **"PC1 explains 40% of the variance"**, we mean:
- PC1 (Principal Component 1) captures 40% of all the "spread" or "movement" in our stocks
- The remaining 60% is captured by other components
- **Higher variance explained = More important!**

### Why We Care About Variance:

- If one component explains 40% of variance, it's VERY important
- It means that component captures a lot of how stocks move
- If a component explains only 1% of variance, it's not very important
- It means that component captures very little of how stocks move

**Think of it like this:**
- Imagine you're explaining why people buy stocks
- If you say "40% of buying is because of tech trends" ‚Üí That's a big factor!
- If you say "1% of buying is because of weather" ‚Üí That's a tiny factor!

## üîç Part 4: What is PCA? (Principal Component Analysis)

### In Plain English:

**PCA** = A way to find the main "patterns" or "themes" in your data

### Simple Analogy:

Imagine you're a teacher with 50 students' test scores in 10 subjects:
- Math, Science, English, History, etc.

**Question:** Can we simplify this? Are there patterns?

**PCA would find:**
- **Pattern 1**: "Good at Math/Science" students (explains 30% of differences)
- **Pattern 2**: "Good at English/History" students (explains 25% of differences)
- **Pattern 3**: "Average at everything" students (explains 15% of differences)
- etc.

Instead of tracking 10 subjects for each student, we track 3-4 patterns!

### In Stock Market Terms:

**PCA finds:**
- **Pattern 1 (PC1)**: "Market factor" - Most stocks move together (explains 40% of movement)
- **Pattern 2 (PC2)**: "Tech vs Energy" - Tech stocks move opposite to energy (explains 15% of movement)
- **Pattern 3 (PC3)**: "Growth vs Value" - Growth stocks move opposite to value (explains 10% of movement)
- etc.

### The Steps of PCA (Simple Version):

1. **Step 1**: Calculate how all stocks move together (correlation matrix)
2. **Step 2**: Find the "main direction" where stocks move the most (PC1)
3. **Step 3**: Find the "second main direction" (PC2)
4. **Step 4**: Keep finding directions until we've captured most of the movement
5. **Step 5**: Each direction is called a "Principal Component" (PC)

### Why Do We Do PCA?

**Benefits:**
1. **Simplify**: 50 stocks ‚Üí 5-10 patterns (easier to understand!)
2. **Find Hidden Patterns**: Discover that tech stocks move together
3. **Reduce Noise**: Focus on what matters (ignore small random movements)
4. **Better Analysis**: Understand the market structure

**Real Example:**
- Instead of saying "Apple, Microsoft, Google, Facebook all moved up"
- We say "The tech factor (PC2) moved up"
- Much simpler!

## üß≠ Part 5: What are Eigenvectors? (Simple Explanation)

### In Plain English:

**Eigenvector** = A "direction" that shows which stocks move together

### Simple Analogy:

Imagine you're in a room with 50 people, and you want to know: "Which way is everyone facing?"

**Eigenvector 1 (PC1)**: "Most people are facing North-East"
- This is the main direction
- Most stocks move in this direction
- This explains the most movement (highest variance)

**Eigenvector 2 (PC2)**: "Some people face North, some face South"
- This is the second main direction
- Some stocks move North, some move South
- This explains the second most movement

### In Stock Terms:

**Eigenvector 1 (PC1)**: 
- Shows which stocks move together in the "market direction"
- Usually, ALL stocks have positive values (they all move with the market)
- This is the "market factor"

**Eigenvector 2 (PC2)**:
- Shows which stocks move together in a "sector direction"
- Tech stocks might have positive values
- Energy stocks might have negative values
- This is the "sector factor"

### What Do the Numbers Mean?

Each stock gets a number for each eigenvector:
- **Positive number** = Stock moves in that direction
- **Negative number** = Stock moves opposite to that direction
- **Large number** = Stock moves strongly in that direction
- **Small number** = Stock doesn't move much in that direction

**Example:**
- Apple: PC1 = +0.15, PC2 = +0.20
  - Moves with market (PC1 positive)
  - Moves with tech sector (PC2 positive, large)
- Oil Company: PC1 = +0.15, PC2 = -0.18
  - Moves with market (PC1 positive)
  - Moves opposite to tech (PC2 negative)

### Eigenvalues (The "Importance" Numbers):

**Eigenvalue** = How important this eigenvector is

- **Large eigenvalue** = This direction explains a lot of movement (important!)
- **Small eigenvalue** = This direction explains little movement (not important)

**Example:**
- PC1 eigenvalue = 15.2 ‚Üí Explains 40% of all movement (VERY important!)
- PC2 eigenvalue = 5.8 ‚Üí Explains 15% of all movement (important)
- PC10 eigenvalue = 0.3 ‚Üí Explains 1% of all movement (not very important)

```python
# Step 2: Fetch historical data using yfinance (parallelized)

from concurrent.futures import ThreadPoolExecutor, as_completed

PERIOD = "1y"  # 1 year of data
INTERVAL = "1h"  # Daily data

returns_dict = {}
failed_symbols = []

def fetch_symbol_return(symbol, idx, total):
    try:
        ticker = yf.Ticker(symbol)
        df = ticker.history(period=PERIOD, interval=INTERVAL)
        if df.empty:
            return (symbol, None, f"‚ùå [{idx:2d}/{total}] {symbol}: No data available")
        returns = np.log(df['Close'] / df['Close'].shift(1)).dropna()
        return (symbol, returns, f"‚úÖ [{idx:2d}/{total}] {symbol}: {len(returns)} trading days")
    except Exception as e:
        return (symbol, None, f"‚ùå [{idx:2d}/{total}] {symbol}: Error - {str(e)[:50]}")

print("üîÑ Fetching data for all stocks using yfinance (parallelized)...")
print("=" * 60)


max_workers = min(55, len(STOCKS))  # You may tune this for your system/network
futures = []
with ThreadPoolExecutor(max_workers=max_workers) as executor:
    for i, symbol in enumerate(STOCKS, 1):
        futures.append(executor.submit(fetch_symbol_return, symbol, i, len(STOCKS)))
    for future in as_completed(futures):
        symbol, returns, msg = future.result()
        print(msg)
        if returns is None:
            failed_symbols.append(symbol)
        else:
            returns_dict[symbol] = returns

print("=" * 60)
print(f"\nüìä Successfully fetched: {len(returns_dict)} stocks")
if failed_symbols:
    print(f"‚ö†Ô∏è  Failed: {len(failed_symbols)} stocks - {', '.join(failed_symbols)}")
```

```python
# Step 3: Align all returns on common dates
returns_df = pd.DataFrame(returns_dict)
returns_df = returns_df.dropna()  # Remove rows with any NaN (missing data)

print("üìê Data Alignment Summary:")
print("=" * 60)
print(f"üìÖ Date range: {returns_df.index[0].date()} to {returns_df.index[-1].date()}")
print(f"üìä Final dataset shape: {returns_df.shape[0]} days √ó {returns_df.shape[1]} stocks")
print(f"üìà Total data points: {returns_df.shape[0] * returns_df.shape[1]:,}")

# Show sample of the data
print("\nüìã Sample of returns data (first 5 days, first 5 stocks):")
print(returns_df.iloc[:5, :5].round(4))
print("\nüí° Each value is the log return: ln(Price_today / Price_yesterday)")
```

## üßÆ Step 6: Understanding Variance - Before We Find Eigenvectors!

### ‚ö†Ô∏è IMPORTANT: What Does "Variance" Mean Here?

Before we find eigenvectors, you need to understand **variance** in this context!

### Simple Explanation of Variance:

**Variance** = How much things "spread out" or "vary"

**Example:**
- Stock A: Price goes $100 ‚Üí $101 ‚Üí $100 ‚Üí $101 (low variance, stable)
- Stock B: Price goes $100 ‚Üí $120 ‚Üí $80 ‚Üí $130 (high variance, volatile)

### In Our Context - "Variance Explained":

When we say **"PC1 explains 40% of the variance"**, here's what it means:

**Think of it like this:**
- Imagine all the stock movements as a big cloud of data points
- **Total variance** = How spread out this entire cloud is
- **PC1 variance** = How much of that spread is in the PC1 direction

**Real Example:**
- Total variance in our 50 stocks = 100% (all the movement)
- PC1 captures 40% of that movement ‚Üí **PC1 explains 40% of variance**
- PC2 captures 15% of that movement ‚Üí **PC2 explains 15% of variance**
- PC3 captures 10% of that movement ‚Üí **PC3 explains 10% of variance**
- Remaining 35% is spread across PC4, PC5, ... PC50

### Why This Matters:

**If PC1 explains 40% of variance:**
- It means PC1 captures 40% of ALL the movement in our stocks
- This is HUGE! It's the most important pattern
- Most stocks move in this direction

**If PC10 explains only 1% of variance:**
- It means PC10 captures only 1% of the movement
- This is small! It's not very important
- Only a few stocks move in this direction

### The Key Insight:

**The first few PCs (PC1, PC2, PC3) explain MOST of the variance!**
- This means we can ignore PC10, PC20, etc. - they don't matter much
- We can simplify 50 stocks to just 5-10 important patterns!

### Now We're Ready!

Now that you understand variance, let's find the eigenvectors and eigenvalues!


```python
# ============================================================================
# STEP 6: Find Eigenvectors and Eigenvalues - THE CORE OF PCA!
# ============================================================================

# What are we doing?
# We're using a mathematical operation called "eigenvalue decomposition"
# This finds the "principal components" - the main patterns in our data

print("=" * 70)
print("üî¨ COMPUTING EIGENVECTORS AND EIGENVALUES")
print("=" * 70)
print("")
print("üí° What's happening:")
print("   - We're analyzing the correlation matrix")
print("   - Finding the main 'directions' where stocks move (eigenvectors)")
print("   - Finding how 'important' each direction is (eigenvalues)")
print("   - This is called 'Principal Component Analysis' (PCA)")
print("")

# Use NumPy's eigenvalue decomposition function
# This does the math for us - don't worry about the details!
eigenvalues, eigenvectors = np.linalg.eigh(corr_matrix)

# Sort in descending order (largest eigenvalue first)
# Why? Larger eigenvalue = More important direction!
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("‚úÖ Decomposition complete!")
print(f"üìä Number of eigenvalues: {len(eigenvalues)}")
print(f"üìä Number of eigenvectors: {eigenvectors.shape[1]}")
print("")

# ============================================================================
# SHOW ALL EIGENVALUES FIRST (not just top 10!)
# ============================================================================

print("=" * 70)
print("üìä ALL EIGENVALUES (sorted by importance)")
print("=" * 70)
print("")
print("üí° Remember:")
print("   - Eigenvalue = How important this direction is")
print("   - Larger eigenvalue = More important")
print("   - We'll calculate variance explained next")
print("")
print(f"{'PC':<6} {'Eigenvalue':<15} {'What it means':<30}")
print("-" * 70)

# Show ALL eigenvalues, not just top 10!
for i, ev in enumerate(eigenvalues, 1):
    importance = "Very Important" if ev > 5 else "Important" if ev > 1 else "Less Important" if ev > 0.5 else "Not Important"
    print(f"PC{i:<5} {ev:<15.4f} {importance}")
    
print("")
print("=" * 70)

# ============================================================================
# CALCULATE VARIANCE EXPLAINED
# ============================================================================

# Proportion of variance explained
# This answers: "What percentage of all movement does each PC capture?"
pve = eigenvalues / eigenvalues.sum() * 100
cumulative_pve = np.cumsum(pve)

print("")
print("=" * 70)
print("üìà VARIANCE EXPLAINED BY EACH PRINCIPAL COMPONENT")
print("=" * 70)
print("")
print("üí° This is the KEY metric!")
print("   - Shows what % of all stock movement each PC captures")
print("   - PC1 usually captures the most (the 'market factor')")
print("   - Later PCs capture less (sector-specific or random movements)")
print("")
print(f"{'PC':<6} {'Eigenvalue':<12} {'Variance %':<12} {'Cumulative %':<15} {'Interpretation':<20}")
print("-" * 70)

# Show ALL PCs with variance explained
for i in range(len(pve)):
    if i < 5:
        interp = "Market Factor" if i == 0 else "Sector Factor" if i == 1 else "Style Factor" if i == 2 else "Other Factor"
    else:
        interp = "Minor Factor"
    
    print(f"PC{i+1:<5} {eigenvalues[i]:<12.4f} {pve[i]:<12.2f}% {cumulative_pve[i]:<15.2f}% {interp}")

print("")
print("=" * 70)
print("üí° KEY INSIGHTS:")
print("=" * 70)

n_80 = np.argmax(cumulative_pve >= 80) + 1
n_90 = np.argmax(cumulative_pve >= 90) + 1
n_95 = np.argmax(cumulative_pve >= 95) + 1 if cumulative_pve[-1] >= 95 else len(eigenvalues)

print(f"   ‚Ä¢ First {n_80} PCs explain 80% of all variance")
print(f"   ‚Ä¢ First {n_90} PCs explain 90% of all variance")
print(f"   ‚Ä¢ First {n_95} PCs explain 95% of all variance")
print("")
print(f"   üéØ This means we can simplify {len(eigenvalues)} stocks to just {n_80}-{n_90} factors!")
print(f"   üéØ Most of the market movement is captured by the first few PCs!")
print(f"   üéØ PC1 is the 'market factor' - when the market moves, most stocks move together")
print("=" * 70)
```

## üéØ Part 7: Understanding Eigenvectors - The "Directions" of the Market

**What are eigenvectors?**
- They are **vectors** (directions) that show which stocks move together
- Each eigenvector is a **linear combination** of stocks
- Stocks with **high positive loadings** move together in that direction
- Stocks with **negative loadings** move opposite

**Why visualize eigenvectors?**
- See which stocks contribute to each principal component
- Understand market structure (tech vs energy vs finance)
- Identify diversification opportunities

## 7. Clustering Analysis

We use hierarchical clustering to identify groups of stocks that move together based on their correlation patterns.


```python
# Hierarchical clustering based on correlation distance
# Convert correlation to distance: distance = 1 - correlation
distance_matrix = 1 - corr_matrix.values

# Perform hierarchical clustering
linkage_matrix = linkage(distance_matrix, method='ward')

# Plot dendrogram
fig, ax = plt.subplots(figsize=(16, 10))
dendrogram(
    linkage_matrix,
    labels=corr_matrix.index,
    leaf_rotation=90,
    leaf_font_size=8,
    ax=ax
)
ax.set_title('Hierarchical Clustering Dendrogram (Ward Linkage)', fontsize=14, fontweight='bold')
ax.set_xlabel('Stock Symbol', fontsize=12)
ax.set_ylabel('Distance', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Extract clusters (e.g., 5 clusters)
n_clusters = 5
cluster_labels = fcluster(linkage_matrix, n_clusters, criterion='maxclust')

# Create cluster assignment DataFrame
cluster_df = pd.DataFrame({
    'Symbol': corr_matrix.index,
    'Cluster': cluster_labels
})

print(f"Cluster assignments ({n_clusters} clusters):")
for cluster_id in range(1, n_clusters + 1):
    stocks_in_cluster = cluster_df[cluster_df['Cluster'] == cluster_id]['Symbol'].tolist()
    print(f"\nCluster {cluster_id} ({len(stocks_in_cluster)} stocks): {', '.join(stocks_in_cluster)}")
```

```python
# Visualize clusters on correlation heatmap
fig, ax = plt.subplots(figsize=(16, 14))

# Reorder correlation matrix by cluster
cluster_order = cluster_df.sort_values('Cluster')['Symbol'].tolist()
corr_ordered = corr_matrix.loc[cluster_order, cluster_order]

sns.heatmap(
    corr_ordered,
    annot=False,
    cmap='coolwarm',
    center=0,
    vmin=-1,
    vmax=1,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

ax.set_title('Correlation Matrix (Ordered by Clusters)', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Visualize clusters on PC1 vs PC2 plot
fig, ax = plt.subplots(figsize=(14, 10))

# Color by cluster
colors_map = plt.cm.tab10
for cluster_id in range(1, n_clusters + 1):
    stocks_in_cluster = cluster_df[cluster_df['Cluster'] == cluster_id]['Symbol'].tolist()
    mask = pc_loadings.index.isin(stocks_in_cluster)
    
    x = pc_loadings.loc[mask, 'PC1']
    y = pc_loadings.loc[mask, 'PC2']
    
    ax.scatter(x, y, s=150, alpha=0.7, label=f'Cluster {cluster_id}', 
               c=[colors_map(cluster_id / n_clusters)], edgecolors='white', linewidth=1.5)
    
    # Annotate
    for symbol in stocks_in_cluster:
        ax.annotate(symbol, (pc_loadings.loc[symbol, 'PC1'], pc_loadings.loc[symbol, 'PC2']), 
                   fontsize=8, alpha=0.8)

ax.axhline(y=0, color='white', linestyle='--', linewidth=1, alpha=0.5)
ax.axvline(x=0, color='white', linestyle='--', linewidth=1, alpha=0.5)
ax.set_xlabel(f'PC1 (Variance: {pve[0]:.2f}%)', fontsize=12)
ax.set_ylabel(f'PC2 (Variance: {pve[1]:.2f}%)', fontsize=12)
ax.set_title('Clusters Visualized on PC1 vs PC2 Space', fontsize=14, fontweight='bold')
ax.legend(loc='best')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 9. Conclusions & Insights

### Key Findings:

1. **Market Structure**: The first few principal components typically explain a large portion of variance, indicating strong common factors (market risk, sector effects).

2. **Diversification**: Stocks with low correlation or negative loadings on the same PC offer better diversification.

3. **Sector Clustering**: Stocks from the same sector often cluster together, showing sector-specific risk factors.

4. **Principal Components Interpretation**:
   - **PC1** (largest eigenvalue): Often represents the "market factor" - most stocks move together
   - **PC2, PC3, etc.**: Represent sector-specific or style factors (growth vs value, tech vs energy, etc.)

5. **Risk Management**: Understanding eigenvector structure helps in:
   - Portfolio optimization (reducing exposure to dominant factors)
   - Risk decomposition (identifying sources of portfolio risk)
   - Factor investing (tilting towards specific eigenvectors)

### Mathematical Summary:

The correlation matrix $\mathbf{C}$ captures pairwise relationships. Its eigenvalue decomposition reveals:

- **Eigenvalues** $\lambda_i$: Variance explained by each principal component
- **Eigenvectors** $\mathbf{v}_i$: Directions of maximum variance (principal components)
- **Clustering**: Stocks with similar eigenvector coordinates move together

This analysis provides a quantitative foundation for understanding market structure and building diversified portfolios.
